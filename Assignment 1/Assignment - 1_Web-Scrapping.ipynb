```python
import requests
URL = "https://www.flipkart.com/search?q=boat+hedphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(page.content,'html5lib')
```

```python
result=soup.find_all(class_="s1Q9rs")
for i in range(0,len(result)):
    result[i]=result[i].text
```

```python
res=soup.find_all(class_="_30jeq3")
for i in range(0,len(res)):
    res[i]=res[i].text[1:]
```

```python
rate=soup.find_all(class_="_3LWZlK")
for i in range(0,len(res)):
    rate[i]=rate[i].text
```

```python
import pandas as pd
d={"Headphone Names":result,"Price":res[:40],"Rating":rate[:40]}
data=pd.DataFrame(d)
```

```python
data.to_csv("data1.csv")
data
```
